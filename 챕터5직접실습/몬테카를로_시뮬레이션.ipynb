```python
import random
import matplotlib.pyplot as plt

random.seed(0) # 결과 비교를 위해 시드 고정

num_points = 100000

#원 안에 들어있는 x,y 좌표
in_x, in_y = [], []

#원 밖에 있는 x, y 좌표
out_x, out_y = [], []

for _ in range(num_points):
    #난수 발생으로 임의의 위치에 점 생성
    x, y = random.uniform(-1.0, 1.0), random.uniform(-1.0, 1.0)

    if x *  x + y * y <= 1.0:
        in_x.append(x)
        in_y.append(y)
    else:
        out_x.append(x)
        out_y.append(y)

#파이값을 계산해보자
# 3.12884
print(4.0 * len(in_x)/num_points)

fig, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect("equal")

ax.scatter(in_x, in_y, c="red", s=0.1)
ax.scatter(out_x, out_y, c="blue", s=0.1)


ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Monte Carlo Pi")

plt.show()
```
