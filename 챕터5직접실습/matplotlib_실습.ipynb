```python
import matplotlib.pyplot as plt 
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

#데이터 가져오기
dates = []
highs = []
lows = []

with open("BTC-USD.csv") as f:
    while line := f.readline():
        splitted = line.split(",")
        dates.append(splitted[0])
        highs.append(splitted[2])
        lows.append(splitted[3])

#컬럼명 삭제하기
del dates[0]
del highs[0]
del lows[0]

#자료 정리
#날짜는 YYYY-MM으로, 가격은 문자열이 아닌 실수형으로 처리
dates = list(map(lambda x:x[:-3], dates))
highs = list(map(float, highs))
lows = list(map(float, lows))

#그래프 그리기
fig, ax = plt.subplots()

ax.plot(dates, highs, label="High")
ax.plot(dates, lows, label="Low")
ax.set_xlabel("Date")
ax.set_ylabel("Stock price")
ax.set_title("Bitcoin USD")

#그래프에 범례를 추가하는 작업을 함
ax.legend()

#가록축에 모든 날짜를 표시하면 복잡해지기 때문에 MultipleLocator을 이용해서 듬성듬성 그려지도록 조정
ax.xaxis.set_major_locator(MultipleLocator(20))

```
